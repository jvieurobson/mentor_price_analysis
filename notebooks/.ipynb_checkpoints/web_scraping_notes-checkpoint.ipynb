A proxy is a server that acts as an intermediary between a client and another server. 
When a client makes a request to a server, the request first goes to the proxy server. 
The proxy server then makes a request to the actual server and returns the response back to the client.

Proxies are useful to web scraping when the website in question has certain limitations on how many times 
a single individual can send requests to a website. 

In general, when building a web scraping project, it is good practice to implement rotating proxies, 
as this reduces the strain on one particular IP address, and reduces the likelihood of the IP address getting banned.

In this particular project, we will be using free proxies provided by the website free-proxy-list.net.

```python
#Import the required libaries. 
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import traceback

#Read list of rotating proxies.
proxies = open(r"../data/rotating_proxies_list.txt", "r")


#Rotating through proxies and returning first responsive proxy.
def test_proxies():
    try:
        for proxy in proxies:
            response = requests.get(f"https://mentorcruise.com/mentor/browse/?search=&page=1", proxies = {'http': f"http://{proxy}"}, timeout = 5)
            status_code = response.status_code
            if status_code == 200:
                print(response.status_code)
                return proxy
    except ConnectionError:
        print("Request blocked by server")


#Create empty list for profile information.
profiles = []

#Loop through pages.        
for page in range(1, 2):
    
    print(page)
    
    #Send requests.
    response = requests.get(f"https://mentorcruise.com/mentor/browse/?search=&page={page}", 
                            proxies = {'http': f"http://{test_proxies}"}, 
                            timeout = 5)    #Make a request.
    
    #Parse HTML.
    soup = BeautifulSoup(response.content, 'html.parser')    #Parse the HTML content of the page. 



    #Loop through profiles and extract feature information.
    for profile in soup.select("div.sm\:grid.grid-cols-24.sm\:space-x-8"):
        
        try:
            name = profile.select_one('h3.title.text-2xl.has-text-blue.font-bold').text.split('\n')[1]
            job_title = profile.select_one('span.has-text-blue.text-base.mt-2.inline-block').text.split('\n')[1].split()[1]
            employer = profile.select_one('span.has-text-blue.text-base.mt-2.inline-block').text.split('\n')[1].split()[1]
            short_description = profile.select_one('div.font-medium.text-teal-700.text-base')
            long_description = profile.select_one('div.break-word.mt-4.mb-6.max-w-screen-md.text-sm.leading-6')
            rating = profile.select_one('b.has-text-blue').text
            num_reviews = profile.select_one('span.rating-display').text.split()[1].replace("(", "")
            services_offered = profile.select_one("#search-results > div:nth-child(1) > div.sm\:grid.grid-cols-24.sm\:space-x-8 > div.col-span-full.col-start-8.md\:col-start-6.lg\:col-span-13.border-0.lg\:border-r.border-solid.border-gray-200.pr-5 > div > div.flex.space-x-4.cursor-default.inline-block.w-full.py-3.my-5.border-0.border-solid.border-t.border-b.border-gray-200.has-text-blue.text-sm.font-semibold").text.split()
            tags = profile.select_one("div.tags").text.split()
            dollars_month = profile.select_one('span.price-element.minimize').text


            #Appending features to profiles list.
            profiles.append({
            'name': name,
            'job_title': job_title,
            'employer': employer,
            'short_description': short_description,
            'long_description': long_description,
            'rating': rating,
            'num_reviews': num_reviews,
            'services_offered': services_offered,
            'tags': tags,
            'dollars_month': dollars_month
        })

        except Exception:
            pass

    if soup.select_one('div.-mt-px.w-0.flex-1.flex.justify-end') is None:
            break

    page +=1
```

## Web Scraping Notes.

Web pages can be represented in multiple different languages. 
The most common languages used are HTML, CSS, and Javascript. HTML is a markup language and is used to provide the basic structure and content of a website. CSS is  a stylesheet language and is used to add visual elements to websites including font size, color, spacing.etc. Javascript is a programming language and is used to build website that are more interactive, dynamic, and repsonsive. 

It is important to note the difference between a markup, stylesheet, and programming language. A markup language is a set of naming conventions used to describe the structure and content of a webpage. A stylesheet language is a set of rules for how the content of a web page should be presented and styled. A programming language is a set of instructions for a computer to perform.

In this, the main difference is in the degree to which each is able to perform complex computations. 

Let us take an example of a HTML document with basic CSS styling.

```python
<!DOCTYPE html>
<html>
  <head>
    <title> This is a new webpage! <title/>
    <style>
      h1 {
        font-size: 36px;
        color: blue;
      }
    </style>
  </head>
  <body>
    <h1>This is a heading</h1>
    <p>This is a paragraph of text.</p>
  </body>
</html>
```

We notice the following in this document.
`<!DOCTYPE html>` . A declaration used at the top of a webpage to indicate to a browser which html version to use to interpret the website. `<!DOCTYPE` is a naming convention used to specify that we are about to declare a document type for our webpage. It is then followed by `html>` to specify exactly what document type and which version of this document type we want a browser to interpret it as. 

A HTML document is them structured according to element. Elements are the building blocks of a HTML document. Everything you see on the page, except for the document declaration at the top, are elements.
